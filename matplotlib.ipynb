```python
days=[1,2,3,4,5,6,7]
sleeping=[8,7,6,7,5,4,9]
study=[4,2,6,1,3,5,5]
workout=[1,2,1,1,2,3,1]
phone=[5,3,2,4,1,3,4]
other=[6,10,9,11,13,9,5]
lbs=['sleeping', 'study','workout','phone','other']
clrs=['m', 'c', 'r', 'k', 'b']
plt.stackplot(days,sleeping, study,workout,phone,other,labels=lbs, colors=clrs)
plt.xlabel("Days")
plt.ylabel("Activities")
plt.title("Your day")
plt.legend()
plt.show()
```

```python
def avg(l):
    s=0
    for i in l:
        s=s+i
        a=s/len(l)
        return a

days=[1,2,3,4,5,6,7]
sleeping=[8,7,6,7,5,4,9]
study=[4,2,6,1,3,5,5]
workout=[1,2,1,1,2,3,1]
phone=[5,3,2,4,1,3,4]
other=[6,10,9,11,13,9,5]
activities=['sleeping', 'study','workout','phone','other']
clrs=['m', 'c', 'r', 'g', 'b']
hrs=[avg(sleeping),avg(study),avg(workout),avg(phone), avg(other)]
plt.pie(hrs,labels=activities, colors=clrs, explode=(0,0,0,0.1,0), autopct='%2.2f%%')
plt.title("Pie chart")
plt.show()
```

```python
import csv
tests=[]
marks=[]
with open('graph_data.txt','r') as file:
    plots=csv.reader(file, delimiter=',')
    for row in plots:
        tests.append(int(row[0]))
        marks.append(int(row[1]))
plt.plot(tests,marks)
plt.xlabel("Tests")
plt.ylabel("Marks")
plt.title("Display Details from file")
plt.show()
```
